## Условие 3: 3 задача

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
plt.figure()
sns.boxplot(x=df['waterfront'], y= df['price'])
plt.xlabel('Вид на море')
plt.ylabel('Цена')
plt.title('Цена относительно вида на море')
```

Стоимость дома с видом на море изначально больше

```python
plt.figure()
sns.boxplot(x=df['floors'], y=df['price'], whis=1.5)

plt.xlabel('Кол-во этажей')
plt.ylabel('Цена')
plt.title('Цена относительно кол-ва этажей')
```

Чем больше этажей, тем больше начальная стоимость. Минимальная цена 2.5 этажей примерно равна 3.5 этажей

```python
plt.figure()
sns.scatterplot(x=df['sqft_lot'], y=df['price'])

plt.title('Цена относительно жилой площади')
plt.xlabel('Площадь')
plt.ylabel('Цена')
```

Дом с самой большой площадью стоит в районе 1-2 миллионов. Может быть обусловлено плохим видом, плохим состоянием и тд

```python
plt.figure()
sns.scatterplot(x=df['condition'], y=df['price'])

plt.title('Цена относительно состояния дома')
plt.xlabel('Состояние')
plt.ylabel('Цена')
```

Дом с состоянием на 4 стоит дороже всего. Возможно, обусловлено большой площадью (опровергается графиком выше), кол-вом комнат, хорошим видом и тд

```python
plt.figure()
sns.scatterplot(x=df['view'], y=df['price'])

plt.title('Цена относительно качества вида')
plt.xlabel('Оценка вида')
plt.ylabel('Цена')
```

Самая дорогая квартира имеет неплохой вид и хорошее состояние, но имеет очень маленькую площадь, по сравнению с другими домами
